```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss, acf, grangercausalitytests
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf,month_plot,quarter_plot
from scipy import signal
import matplotlib.pyplot as plt
import seaborn as sns 
import date_interpreter
import datetime
from statsmodels.tsa.seasonal import seasonal_decompose
%matplotlib inline 
sns.set_style("whitegrid")
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
```

```python
raw_nope=pd.read_csv(r'Data/Production_Data.csv',index_col=0)
nope = pd.read_csv(r'Data/Production_Data.csv',index_col=0)
date_interpreter.interpret_datestr(nope)
```

```python
nope = nope.set_index('Datetime')
```

```python
nope.head()
```

```python
mean_daily_nope = nope.groupby(['Time']).mean()
```

```python
mean_daily_nope['nope_sma5'] = mean_daily_nope['NOPE'].rolling(5,min_periods=1).mean()
mean_daily_nope['std5'] = mean_daily_nope['NOPE'].rolling(5,min_periods=1).std()
mean_daily_nope['nope_ema'] = mean_daily_nope['NOPE'].ewm(alpha =0.1, adjust = False).mean()
```

```python
ax = plt.figure(figsize=(12,8))
x =mean_daily_nope.index
plt.plot(x, mean_daily_nope['NOPE'])
plt.plot(x, mean_daily_nope['nope_sma5'], lw = 3)
plt.plot(x, mean_daily_nope['nope_ema'], lw = 3)
plt.fill_between(x, mean_daily_nope['nope_sma5']-mean_daily_nope['std5'], mean_daily_nope['nope_sma5']+mean_daily_nope['std5'], color = 'red', alpha= 0.3)
plt.title('Daily NOPE Behavior')
plt.ylabel('NOPE')
plt.legend(['NOPE','SMA-5','EMA','1 Std Dev'])
```
